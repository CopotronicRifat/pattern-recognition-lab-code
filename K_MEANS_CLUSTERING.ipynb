```
# dataset

import matplotlib.pyplot as plt
from random import random

x = []
y = []

for i in range(50):
  x.append(random())
  y.append(random())

for i in range(10):
  x.append(random())
  y.append(-random())

for i in range(10):
  x.append(-random())
  y.append(random())

for i in range(50):
  x.append(-random())
  y.append(-random())

c1x = random()
c1y = random()
c2x = -random()
c2y = -random()

plt.scatter(c1x, c1y, marker = "x", color = "red")
plt.scatter(c2x, c2y, marker = "x", color = "green")
plt.scatter(x, y)
plt.show()

for i in range(10):

  # distance
  d1 = []
  for i in range(len(x)):
    temp1 = (c1x - x[i])**2 + (c1y - y[i])**2
    d1.append(temp1)

  d2 = []
  for i in range(len(x)):
    temp2 = (c2x - x[i])**2 + (c2y - y[i])**2
    d2.append(temp2)


  sum1x = 0
  sum1y = 0
  sum2x = 0
  sum2y = 0
  red = 0
  green = 0
  for i in range(len(x)):
    if d1[i]<d2[i]:
      plt.scatter(x[i], y[i], color = "red")
      sum1x = sum1x + x[i]
      sum1y = sum1y + y[i]
      red = red + 1
    else:
      plt.scatter(x[i], y[i], color = "green")
      sum2x = sum2x + x[i]
      sum2y = sum2y + y[i]
      green = green + 1

  c1x = sum1x / red
  c1y = sum1y / red
  c2x = sum2x / green
  c2y = sum2y / green

  plt.scatter(c1x, c1y, marker = "x", color = "red")
  plt.scatter(c2x, c2y, marker = "x", color = "green")
  plt.show()
```
