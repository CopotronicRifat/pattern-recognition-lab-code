```
import io
import pandas as pd
df2 = pd.read_csv(io.BytesIO(uploaded['DATA.csv']))
```

```
print(df2)
x = df2['AREA']
print(x[0])
y = df2['PRICE']


print(x)
print(y)
```

```

import matplotlib.pyplot as plt

# parameter initialization
theta = [2, 2]
J = 100
n = 0 
m = len(x)
alpha = 0.01
for i in range(100):
  print("Iteration number: ",n+1)
  n = n+1
  # hypothesis function
  h = []
  for i in range(m):
    temp = theta[0] + theta[1]*x[i]
    h.append(temp)
  print("Hypothesis function value is: ",h)

  # cost function
  error_sum = 0 
  for i in range(m):
    error_sum = error_sum + (h[i] - y[i])**2

  J = (1/(2*m))*error_sum

  print("Cost function is:",J)

  # gradient descent
  temp0 = 0
  for i in range(m):
    temp0 = temp0 + (h[i] - y[i])

  theta[0] = theta[0] - (alpha/m)*temp0

  temp1 = 0
  for i in range(m):
    temp1 = temp1 + (h[i] - y[i])*x[i]

  theta[1] = theta[1] - (alpha/m)*temp1

  print("New parameter value is: ",theta)
  plt.plot(x, y, marker = "x")
  
  plt.plot(x, h, marker = "o")
  plt.show()
  

print("predicted hypothesis is:",h)
```
