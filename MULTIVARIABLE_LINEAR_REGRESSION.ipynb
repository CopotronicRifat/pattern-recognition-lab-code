```python
# dataset
import os
import io
import pandas as pd

df = pd.read_csv('MULTI DATA.csv')
print(df)


x1 = df["x1"]
x2 = df["x2"]
x3 = df["x3"]
y = df["y"]
```

```python
# parameter initialization
theta = [1, 1.5, 2, 1.7]

print(theta)
```

```python
J = 100
theta = [1, 1.5, 2, 1.7]
for i in range(10000):
    # hypothesis function
    
    h = []
    m = len(y)
    for i in range(m):
        h.append(theta[0] + theta[1]*x1[i] + theta[2]*x2[i] + theta[3]*x3[i])

    print("Hypothesis Function Value is: ", h)

    # cost

    error = 0
    for i in range(m):
        error = error + (h[i] - y[i]) ** 2

    J = (1/ (2*m))*error

    print("Cost function value is", J)


    # Gradient Descent
    sum0 = 0
    sum1 = 0
    sum2 = 0
    sum3 = 0
    for i in range(m):
        sum0 = sum0 + (h[i] - y[i])
        sum1 = sum1 + (h[i] - y[1])*x1[i]
        sum2 = sum2 + (h[i] - y[i])*x2[i]
        sum3 = sum3 + (h[i] - y[i])*x3[i]


    alpha = 0.001
    theta[0] = theta[0] - (alpha/m)*sum0
    theta[1] = theta[1] - (alpha/m)*sum1
    theta[2] = theta[2] - (alpha/m)*sum2
    theta[3] = theta[3] - (alpha/m)*sum3

    print("Updated parameters are: ", theta)
    
```
